Below are several examples, generated by Google's Colab AI, of f-string syntax.

```python
# Simple f-string example
name = "Eric Idle"
print(f"Hello, {name}")

# Example with expressions
age = 74
print(f"Hello, {name}. You are {age}.")

# Example with function calls
import math
radius = 10
area = math.pi * radius**2
print(f"The area of a circle with radius {radius} is {area:.2f}.")

# Example with string formatting
first_name = "John"
last_name = "Cleese"
age = 80
print(f"{first_name} {last_name} is {age}.")

# Example with dictionary unpacking
person = {"first_name": "Michael", "last_name": "Palin", "age": 79}
print(f"{person['first_name']} {person['last_name']} is {person['age']}.")
```
